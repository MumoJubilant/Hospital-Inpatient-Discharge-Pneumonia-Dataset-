```python
# Identify the numeric columns
numeric_columns = df_1.select_dtypes(include=['number']).columns

# Print numeric columns for verification
print("Numeric Columns:", numeric_columns)
```

```python
categorical_columns = df_1.select_dtypes(include=['object']).columns

# Print numeric columns for verification
print("Categorical Columns:", categorical_columns)
```

```python
# Fill missing values in categorical columns with the mode
df_1[categorical_columns] = df_1[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Display the first few rows of the updated DataFrame
df_1.head()
```

```python
# Define common placeholders
placeholders = ["N/A", "NA", "null", "None", "undefined", "", 'nan']

# Initialize a dictionary to hold placeholder counts
placeholder_counts = {placeholder: 0 for placeholder in placeholders}

# Loop through each placeholder and count its occurrences
for placeholder in placeholders:
    # Use case-insensitive comparison and strip leading/trailing spaces
    placeholder_count = (data_1.applymap(lambda x: str(x).strip().lower()) == placeholder.lower()).sum().sum()
    if placeholder_count > 0:
        placeholder_counts[placeholder] = placeholder_count

# Print the counts of each placeholder
for placeholder, count in placeholder_counts.items():
    if count >= 0:
        print(f"Found {count} occurrences of placeholder '{placeholder}'")
```

HANDLING OUTLIERS

```python
#Lets first check if there are any outliers
numeric_columns = data_1.select_dtypes(include=['float64', 'int64', 'int32'])

# Plot box plots for each numeric column
num_cols = len(numeric_columns.columns)
cols_per_row = 3
num_rows = (num_cols - 1) // cols_per_row + 1
plt.figure(figsize=(15, 5 * num_rows))

for i, col in enumerate(numeric_columns.columns):
    plt.subplot(num_rows, cols_per_row, i+1)
    plt.boxplot(numeric_columns[col].dropna())  # Drop NaN values for plotting
    plt.title(col)

plt.tight_layout()
plt.show()
```

There seems to be outliers in the total costs, APR MDC Code, APR Severity of illness code, Total Charges, APR DRG Code, CCS Procedure Code columns. Lets remove outliers with IQR method.

```python
# Columns with potential outliers
columns_with_outliers = ['Total Costs', 'APR MDC Code', 'APR Severity of Illness Code', 'Total Charges', 'APR DRG Code', 'CCS Procedure Code']

# Function to remove outliers using IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Apply the function to each column with potential outliers
for column in columns_with_outliers:
    data_1 = remove_outliers_iqr(data_1, column)

# Display the first few rows of the updated DataFrame
data_1.head()
```

```python
#lets check if we have cleared the outliers
numeric_columns = df_1.select_dtypes(include=['float64', 'int64', 'int32'])

# Plot box plots for each numeric column
num_cols = len(numeric_columns.columns)
cols_per_row = 3
num_rows = (num_cols - 1) // cols_per_row + 1
plt.figure(figsize=(15, 5 * num_rows))

for i, col in enumerate(numeric_columns.columns):
    plt.subplot(num_rows, cols_per_row, i+1)
    plt.boxplot(numeric_columns[col].dropna())  # Drop NaN values for plotting
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
#Checking for white spaces
def check_white_spaces(df):
    columns_with_whitespaces = []
    for column in df.select_dtypes(include=['object']).columns:
        if df[column].apply(lambda x: isinstance(x, str) and (x != x.strip())).any():
            columns_with_whitespaces.append(column)
    return columns_with_whitespaces

# List columns with white spaces
columns_with_whitespaces = check_white_spaces(data_1)
print("Columns with leading or trailing white spaces:", columns_with_whitespaces)
```
